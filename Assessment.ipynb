```python
import json

input_file = "/content/Chat Data for assessment of applicants.json"
output_file = "/content/repaired_dataset.json"

with open(input_file, "r", encoding="utf-8") as f:
    text = f.read()

decoder = json.JSONDecoder()

objs = []
idx = 0

while idx < len(text):
    # Skip whitespace and commas between objects
    while idx < len(text) and (text[idx].isspace() or text[idx] == ","):
        idx += 1

    if idx >= len(text):
        break

    try:
        obj, end = decoder.raw_decode(text, idx)
        objs.append(obj)
        idx = end
    except json.JSONDecodeError:
        idx += 1

print("Recovered conversations:", len(objs))

with open(output_file, "w", encoding="utf-8") as f:
    json.dump(objs, f, ensure_ascii=False, indent=2)

print("Saved:", output_file)
```

```python
import json

with open("/content/repaired_dataset.json", "r", encoding="utf-8") as f:
    data = json.load(f)

print("Total conversations:", len(data))
```

```python
import pprint

pprint.pprint(data[0])
```

```python
from transformers import AutoTokenizer

model_name = "Qwen/Qwen2.5-3B-Instruct"

tokenizer = AutoTokenizer.from_pretrained(model_name)
```

```python
texts = []

for conv in data:

    chat = tokenizer.apply_chat_template(
        conv["messages"],
        tokenize=False,
        add_generation_prompt=False
    )

    texts.append({"text": chat})

print(texts[0]["text"][:1000])
```

```python
from datasets import Dataset

dataset = Dataset.from_list(texts)

dataset
```

```python
lengths = []

for t in texts:
    ids = tokenizer(t["text"]).input_ids
    lengths.append(len(ids))

print("Max tokens:", max(lengths))
print("Average:", sum(lengths)/len(lengths))
```

```python
import torch
from transformers import AutoModelForCausalLM, BitsAndBytesConfig

# Define the quantization configuration
bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_quant_type="nf4", # recommended from HuggingFace docs for 4-bit
    bnb_4bit_compute_dtype=torch.bfloat16, # bfloat16 is generally good for newer GPUs and better numerical stability
    bnb_4bit_use_double_quant=False,
)

model = AutoModelForCausalLM.from_pretrained(
    model_name,
    device_map="auto",
    quantization_config=bnb_config
)
```

```python
from peft import LoraConfig

peft_config = LoraConfig(
    r=16,
    lora_alpha=32,
    target_modules=[
        "q_proj",
        "k_proj",
        "v_proj",
        "o_proj"
    ],
    lora_dropout=0.05,
    bias="none",
    task_type="CAUSAL_LM"
)
```

```python
from transformers import TrainingArguments

training_args = TrainingArguments(
    output_dir="./qwen_astrologer",
    per_device_train_batch_size=1,
    gradient_accumulation_steps=8,
    learning_rate=1e-4,
    num_train_epochs=8,
    logging_steps=1,
    save_strategy="epoch",
    fp16=True,
    report_to="none",
    warmup_ratio=0.1,
    weight_decay=0.01
)
```

```python
from datasets import Dataset

dataset = Dataset.from_list(texts)

dataset = dataset.train_test_split(
    test_size=0.15,
    seed=42
)

train_dataset = dataset["train"]
eval_dataset = dataset["test"]

print(train_dataset)
print(eval_dataset)
```

```python
from trl import SFTTrainer

trainer = SFTTrainer(
    model=model,
    train_dataset=train_dataset, # Use the specific training dataset
    eval_dataset=eval_dataset,   # Add the evaluation dataset
    peft_config=peft_config,
    args=training_args,
)
```

```python
from transformers import TrainingArguments
from trl import SFTTrainer

# Re-define TrainingArguments with bf16=True and fp16=False
training_args = TrainingArguments(
    output_dir="./qwen_astrologer",
    per_device_train_batch_size=1,
    gradient_accumulation_steps=8,
    learning_rate=1e-4,
    num_train_epochs=8,
    logging_steps=1,
    save_strategy="epoch",
    fp16=False, # Disable fp16
    bf16=True,  # Enable bf16 for bfloat16 computation
    report_to="none",
    warmup_ratio=0.1,
    weight_decay=0.01
)

# Re-initialize trainer with the corrected training_args
trainer = SFTTrainer(
    model=model,
    train_dataset=train_dataset,
    eval_dataset=eval_dataset,
    peft_config=peft_config,
    args=training_args,
)

trainer.train()
```

```python
metrics = trainer.evaluate()
print(metrics)
```

```python
test_prompts = [
    "Meri job kab lagegi?",
    "Meri shaadi kab hogi?",
    "Business mein loss ho raha hai.",
    "Mera breakup ho gaya hai.",
    "Mujhe foreign travel ke chances bataiye."
]
```

```python
from transformers import pipeline

pipe = pipeline(
    "text-generation",
    model=model,          # your fine-tuned model
    tokenizer=tokenizer
)

for i, prompt in enumerate(test_prompts):

    messages = [
        {
            "role": "system",
            "content": "You are Vedaz's AI Vedic astrologer."
        },
        {
            "role": "user",
            "content": prompt
        }
    ]

    text = tokenizer.apply_chat_template(
        messages,
        tokenize=False,
        add_generation_prompt=True
    )

    output = pipe(
        text,
        max_new_tokens=200,
        temperature=0.7,
        do_sample=True
    )

    print("="*80)
    print("Prompt:", prompt)
    print()
    print(output[0]["generated_text"])
    print("="*80)
```
